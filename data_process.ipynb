```python
import pandas as pd
from PIL import Image
from io import BytesIO
from IPython.display import display
import random
random.seed(10)
df =  pd.read_parquet('simpsons_original.parquet')
for i in range(len(df)):
  img=Image.open(BytesIO(df.iloc[i][0]['bytes']))
  r=random.uniform(0, 1)
  if r<0.2:
    img=img.convert('L')
    img.save("dataset/test/img_{}.jpg".format(i))
  elif r<0.4:
    img.save("dataset/validation/img_{}.jpg".format(i))
  else:
    img.save("dataset/train/img_{}.jpg".format(i))
```
